## Đọc và hiển thị img

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("Felis_silvestris_silvestris_Luc_Viatour.jpg",0)

plt.imshow(img,cmap = 'gray')
plt.show()
```

## Tăng/ giảm độ sáng

```python
img = cv2.imread("Felis_silvestris_silvestris_Luc_Viatour.jpg")

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
import numpy as np


new_img = img.astype(np.float32) + 0
new_img = np.clip(new_img ,0,255)
new_img = new_img.astype ( np.uint8)

plt.imshow(new_img)
```
