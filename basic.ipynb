# switch case

```python
time = str(input())

if time == "5 AM":
    print("Wake up")
elif time == "6 AM":
    print("Yoga")
elif time == "7 AM":
    print("Work")
else:
    print("Do something else")
```

```python
time = str(input())

match time:
    case "5 AM":
        print("Wake up")
    case "6 AM":
        print("Yoga")
    case "7 AM":
        print("Work")
    case _:
        print("Do something else")
```

# Enumerate

```python
list = ['carot','apple','milk']

for i in range (len(list)):
    print(f"{i + 1}. {list[i].upper()} ")
```

```python
for x, y in enumerate(list, start=1):
    print(f"{x}. {y.upper()} ")
```

```python
food_list = [
            ['bo','pizza','sua'],
            ['xuc xich','tao','kem'],
            ['carot','banh dau','cupcake']
            ]

search_item = ['carot','tao','sua']
for i in range (len(food_list)):
    for j in range (len(food_list[i])):
        if food_list[i][j] in search_item:
            print(f"{food_list[i][j]} duoc tim"
                  f'thay o hang {i +1} va cot {j+1}')
```
